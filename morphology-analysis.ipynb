Qualitative tests of normality assumption, almost all spot distributions are normal

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/morphology/nuclei_morphology.csv')
```

```python
for j in range(38):
    fig, ax = plt.subplots(1, 1)
    ax.hist(df[df.columns[j+1]], bins=60)
    ax.title.set_text(df.columns[j+1])
    ax.set_ylabel('Number of Spots')
    plt.show()
    
```
